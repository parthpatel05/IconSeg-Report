```python
import numpy as np
import matplotlib.pyplot as plt

# Load the .npy files
index = 2
data1 = np.load(f'cityscapes/train/depth/{index}.npy')
data2 = np.load(f'cityscapes/train/image/{index}.npy')
data3 = np.load(f'cityscapes/train/label/{index}.npy')

# Create subplots to display images side by side
def viewer(depthIm, rgbIm, maskIm):

    fig, axs = plt.subplots(1, 3, figsize=(15, 5))  # 1 row, 3 columns

    # Display each image
    axs[0].imshow(depthIm)
    axs[0].set_title('Depth')

    axs[1].imshow(rgbIm)
    axs[1].set_title('RGB')

    axs[2].imshow(maskIm)
    axs[2].set_title('Mask')

    # Remove axes for a cleaner look
    for ax in axs:
        ax.axis('off')

    # Show the images
    plt.show()

viewer(data1, data2, data3)
```
